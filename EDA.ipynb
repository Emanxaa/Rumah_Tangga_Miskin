```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loop tiap kolom
for col in df.columns:
    # Hitung jumlah kemunculan tiap label pada kolom
    counts = df[col].value_counts()

    # Buat plot
    plt.figure(figsize=(6,4))
    sns.barplot(x=counts.index, y=counts.values)
    plt.title(f"Distribusi label pada kolom: {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import math

# Tentukan jumlah kolom
n_cols = df.shape[1]

# Atur grid layout (misal 3 kolom per baris)
ncols = 3
nrows = math.ceil(n_cols / ncols)

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*5, nrows*4))
axes = axes.flatten()  # supaya mudah diakses sebagai list

for i, col in enumerate(df.columns):
    counts = df[col].value_counts()
    sns.barplot(x=counts.index, y=counts.values, ax=axes[i])
    axes[i].set_title(f"{col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=45)

# Jika subplot lebih banyak dari jumlah kolom df, sembunyikan sisanya
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
for col in df.columns:
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
for col in df.select_dtypes(include='number').columns:  # hanya kolom numerik
    plt.figure(figsize=(5,4))
    plt.hist(df[col], bins=30, edgecolor="black")
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
num_cols = df.select_dtypes(include='number').columns
n_cols = len(num_cols)

ncols = 3   # jumlah kolom grid
nrows = math.ceil(n_cols / ncols)

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*5, nrows*4))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    axes[i].hist(df[col], bins=30, edgecolor="black")
    axes[i].set_title(f"{col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

# sembunyikan subplot kosong
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```
